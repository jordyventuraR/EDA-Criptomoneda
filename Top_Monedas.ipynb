```python
import pandas as pd
```

```python
monedas = pd.read_csv('monedas_combinadas.csv')
monedas.info()
```

```python
monedas.head()
```

Media de las monedas

```python
#Agrupa dos conjuntos en categoria de uno y saca el promedio por ejemplo
#Agrupa por generos y saca la nota promedio
monedas_operador = monedas[["High", "Low", "Close", "Volume", "moneda"]]
monedas_operador.head()
```

```python
monedas_resumen = monedas_operador.groupby(by="moneda").mean()
monedas_resumen.head()
```

```python
#Monedas con mayor numero de transacciones
monedas_transacciones_plus = monedas_resumen.sort_values(by=['Volume'], ascending=False)
mas_transacciones =  monedas_transacciones_plus.head(10)

```

```python
#Monedas con menor numero de transacciones
monedas_transacciones_minus = monedas_resumen.sort_values(by=['Volume'], ascending=True)
menos_transacciones = monedas_transacciones_minus.head(10)
```

Las medias mayores


```python
monedas_resumen['promedio'] = (monedas_resumen['High'] + monedas_resumen['Low'])/2
monedas_resumen['Rango'] = (monedas_resumen['High'] - monedas_resumen['Low'])
monedas_resumen.head()
```

```python
valor_moneda = monedas_resumen["promedio"]
valor_moneda_asc = valor_moneda.sort_values(ascending=True)
monedas_menor_valor =  valor_moneda_asc.head(10)
menor_valor = monedas_menor_valor.head(10)
```

```python
valor_moneda_des = valor_moneda.sort_values(ascending=False)
monedas_mayor_valor =  valor_moneda_des.head(10)
mayor_valor = monedas_mayor_valor.head(10)
```

```python
valor_rango = monedas_resumen["Rango"]
valor_rango_asc = valor_rango.sort_values(ascending=False)
monedas_mayor_rango =  valor_rango_asc.head(10)
rango_mayor = monedas_mayor_rango.head(10)
```

```python
valor_rango_des = valor_rango.sort_values(ascending=True)
monedas_menor_rango =  valor_rango_des.head(10)
menor_rango = monedas_menor_rango.head(10)
```

```python

media_prom  = monedas_resumen["promedio"].mean()
media_rango = monedas_resumen["Rango"].mean()
print(media_prom)
print(media_rango)
```

```python
#monedas_resumen[["columna1", "columna2"]].describe()

monedas_resumen.describe()
```

```python
arriba_med_prome = monedas_resumen[monedas_resumen["promedio"] > media_prom]
arriba_med_rango = monedas_resumen[monedas_resumen["Rango"] > media_prom]
abajo_med_prome = monedas_resumen[monedas_resumen["promedio"] < media_prom]
abajo_med_rango = monedas_resumen[monedas_resumen["Rango"] < media_prom]
```
