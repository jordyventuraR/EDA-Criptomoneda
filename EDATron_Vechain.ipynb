```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
dataTron2 = pd.read_csv('/content/vechain.csv')


if 'price' in data.columns:
    # Calcular la media de la columna 'price'
    mean_price = data['price'].mean()

    # Graficar la densidad de la columna 'price' usando pandas
    ax = data['price'].plot.density(title="Curva de Densidad de Precios")

    # Agregar una línea vertical en la media
    plt.axvline(mean_price, color='red', linestyle='dashed', linewidth=2, label=f'Media: {mean_price:.2f}')

    # Agregar una leyenda
    plt.legend()

    # Mostrar el gráfico
    plt.show()
else:
    print("La columna 'price' no existe en el archivo CSV.")

```

```python
import matplotlib.pyplot as plt

# Definir los valores de la media y la desviación estándar
mean_value = 0.033532
std_value = 0.040655

# Crear un gráfico de dispersión (scatter plot)
plt.scatter(mean_value, std_value, color='blue', label=f'Media: {mean_value}, STD: {std_value}')
plt.title("Gráfico de Dispersión entre Media y Desviación Estándar")
plt.xlabel("Media")
plt.ylabel("Desviación Estándar")
plt.legend()
plt.grid(True)
plt.show()
```
