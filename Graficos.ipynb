```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
monedas_estables = pd.read_csv('rangomenor.csv')

# Configurar el gr치fico
plt.figure(figsize=(10, 6))  # Ajustar el tama침o del gr치fico
monedas_estables.plot(x="moneda", y="Rango", kind="bar", legend=False)

plt.title("Monedas m치s estables")
plt.xlabel("Monedas")
plt.ylabel("Rango")
plt.grid()


plt.show()
```

```python
tron = pd.read_csv('tron.csv')
tron.head()
precio_tron = tron[['date', 'price']]
precio_tron.head()

# Graficar
precio_tron.plot(x="date", y="price", kind="line", marker='o')
plt.title("Moneda Tron")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.xticks(rotation=45)  # Rotar etiquetas del eje X
plt.grid()

# Mostrar el gr치fico
plt.show()
```
